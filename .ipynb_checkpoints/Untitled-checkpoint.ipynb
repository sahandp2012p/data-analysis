```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from wordcloud import WordCloud
```

```python
videos = pd.read_csv("US_youtube_trending_data.csv")
category_ids = pd.read_csv("category_ids.csv")

videos.view_count.max()
```

```python
videos.drop(["video_id", "channelId", "thumbnail_link",
         "comments_disabled", "ratings_disabled"], axis=1, inplace=True)
```

```python
published = []

for i in range(len(videos.publishedAt)):
    published.append(pd.to_datetime(videos.publishedAt[i]).strftime("%b"))
    
published = pd.Series(published)

videos["month_published"] = published
```

```python
trended = []

for j in range(len(videos.trending_date)):
    trended.append(pd.to_datetime(videos.trending_date[j]).strftime("%b"))
    
trended = pd.Series(trended)
    
videos["month_trending"] = trended
```

```python
lags = []

for k in range(len(videos)):
        lags.append(pd.to_datetime(videos.trending_date[k]) - pd.to_datetime(videos.publishedAt[k]))

lags = pd.Series(lags)
                    
videos["lag"] = lags
```

```python
categories = videos.categoryId

for categoryId in videos.categoryId.unique():
    categories.replace(categoryId, category_ids.loc[category_ids.categoryId
                                                    == categoryId].values[0][1], inplace=True)
```

```python
lags = []

for k in range(len(videos)):
        lags.append(pd.to_datetime(videos.trending_date[k]) - pd.to_datetime(videos.publishedAt[k]))

lags = pd.Series(lags)
                    
videos["lag"] = lags
```

```python
videos.rename(columns={"channelTitle": "channel", "view_count": "views",
                       "categoryId": "category", "comment_count": "comments"}, inplace=True)
```

```python
videos.drop_duplicates(subset="title", keep="first", inplace=True)
```

```python
videos.sort_values(by="likes", ascending=False, inplace=True)
```

```python
mon = 0
tue = 0
wed = 0
thu = 0
fri = 0
sat = 0
sun = 0

def check(index):
    return pd.to_datetime(videos.iloc[l].publishedAt).strftime("%a")

for l in range(len(videos)):
    if check(l) == 'Mon':
        mon += 1
    elif check(l) == 'Tue':
        tue += 1
    elif check(l) == 'Wed':
        wed += 1
    elif check(l) == 'Thu':
        thu += 1
    elif check(l) == 'Fri':
        fri += 1
    elif check(l) == 'Sat':
        sat += 1
    else:
        sun += 1
```

```python
fig, ax = plt.subplots()

ax.scatter("Mon", mon, label="Mon")
ax.scatter("Tue", tue, label="Tue", color="pink")
ax.scatter("Wed", wed, label="Wed", color="indigo")
ax.scatter("Thu", thu, label="Thu", color="green")
ax.scatter("Fri", fri, label="Fri", color="orange")
ax.scatter("Sat", sat, label="Sat", color="limegreen")
ax.scatter("Sun", sun, label="Sun", color="red")

ax.set_title("Videos uploaded in each day")
ax.set_xlabel("Day of the week")
ax.set_ylabel("Videos uploaded in the day")

ax.legend(frameon=True)
```

```python
lag_data = videos.lag.dt.days
plt.hist(lag_data, density=True, bins=30)

lag_data.max()
```

```python
data = videos[["month_published", "likes"]]
    
likes = [like for like in data.groupby(["month_published"]).likes.sum()]

august = len(data[videos.month_trending == "Aug"])
september = len(data[videos.month_trending == "Sep"])
october = len(data[videos.month_trending == "Oct"])
november = len(data[videos.month_trending == "Nov"])

fig, ax = plt.subplots()

ax.set_title("Videos for each Month")

ax.set_xlabel("Month")
ax.set_ylabel("Videos")

ax.bar(["August", "September", "October", "November"], [august, september, october, november], label="Videos")

ax.legend(frameon=True)
```

```python
channels =  videos.sort_values(by="month_trending", ascending=False)[["channel", "views", "month_trending"]]

channels = channels.sort_values(by="views", ascending=False)

top10_aug = channels.loc[channels.month_trending == "Aug"].groupby("channel").views.sum().sort_values(
    ascending=False).head(10)

top10_sep = channels.loc[channels.month_trending == "Sep"].groupby("channel").views.sum().sort_values(
    ascending=False).head(10)

top10_oct = channels.loc[channels.month_trending == "Oct"].groupby("channel").views.sum().sort_values(
    ascending=False).head(10)

top10_nov = channels.loc[channels.month_trending == "Nov"].groupby("channel").views.sum().sort_values(
    ascending=False).head(10)
```

```python
fig, ax = plt.subplots()

plt.xticks(rotation=90)

ax.set_title("Top 10 Trending Channels for August")

ax.set_ylabel("Views by Millions")

ax.bar(top10_aug.index, top10_aug, label="Views")

ax.legend(frameon=True)
```

```python
fig, ax = plt.subplots()

plt.xticks(rotation=90)

ax.set_title("Top 10 Trending Channels for September")

ax.set_ylabel("Views by Millions")

ax.bar(top10_sep.index, top10_sep, label="Views")

ax.legend(frameon=True)
```

```python
fig, ax = plt.subplots()

plt.xticks(rotation=90)

ax.set_title("Top 10 Trending Channels for October")

ax.set_ylabel("Views by Millions")

ax.bar(top10_oct.index, top10_oct, label="Views")

ax.legend(frameon=True)
```

```python
fig, ax = plt.subplots()

plt.xticks(rotation=90)

ax.set_title("Top 10 Trending Channels for November")

ax.set_ylabel("Views by Millions")

ax.bar(top10_nov.index, top10_nov, label="Views")

ax.legend(frameon=True)
```

```python
videos["total_opinions"] = videos.likes + videos.dislikes

videos["%likes"] = (videos.likes / videos.total_opinions) * 100
videos["%dislikes"] = (videos.dislikes / videos.total_opinions) * 100
```

```python
plt.xticks(rotation=90)

plt.bar(videos.category.value_counts().index, videos.category.value_counts())
```

```python
pubs = []

for pub in videos.publishedAt:
    if pd.to_datetime(pub).strftime("%x") == "08/15/20":
        pubs.append(videos[videos.publishedAt == pub])
        
fig, ax = plt.subplots()  

ax.set_title("Views 08/15/2020")
ax.set_xlabel("Views")
ax.set_ylabel("Reactions")

for video in pubs:
    ax.scatter(video.views, video.likes, c="#fcca05")
    ax.scatter(video.views, video.dislikes, c="#8205ff")
    ax.scatter(video.views, video.comments, c="#05afff")  
    
ax.legend(labels=["Likes", "Dislikes", "Comments"])
```
